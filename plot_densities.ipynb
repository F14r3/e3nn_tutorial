```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from functools import partial
```

```python
basis = lambda r, alpha, norm : norm*np.exp(-alpha*r**2)

alpha = np.abs(np.random.randn(3, 1))
norm = np.abs(np.random.randn(3,1))

basis_on_r = partial(basis, alpha=alpha, norm=norm)
```

```python
x = np.linspace(0, 5, 100)
y = basis_on_r(x)

for i in range(3):
    plt.plot(x,y[i])
    plt.show
```
